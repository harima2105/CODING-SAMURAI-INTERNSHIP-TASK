```python
#Import data manipulation library
import pandas as pd
#Import vizualization library
import matplotlib.pyplot as plt
```

```python
#Load the dataset
df = pd.read_csv("IBM HR-Employee-Attrition.csv")
```

```python
#View basic info
print("Dataset Overview:")
print(df.head())
```

```python
#Calculate Attrition Rate
total_employees = len(df)
attrition_count = df[df['Attrition'] == 'Yes'].shape[0]
attrition_rate = (attrition_count / total_employees) * 100
print(f"\nAttrition Rate: {attrition_rate:.2f}%")
```

```python
#Visualize Attrition by Department
attrition_by_dept = df[df['Attrition'] == 'Yes']['Department'].value_counts()
plt.figure(figsize=(8,5))
attrition_by_dept.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Attrition by Department', fontsize=14)
plt.xlabel('Department')
plt.ylabel('Number of Employees Left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average Age of Employees Who Left
avg_age_left = df[df['Attrition'] == 'Yes']['Age'].mean()
print(f"\nAverage Age of Employees Who Left: {avg_age_left:.1f} years")
```

```python
summary = {
    "Total Employees": total_employees,
    "Employees Left": attrition_count,
    "Attrition Rate (%)": round(attrition_rate, 2),
    "Average Age of Employees Left": round(avg_age_left, 1)
}
summary_df = pd.DataFrame([summary])
summary_df.to_csv("Attrition_Summary.csv", index=False)
print("\nSummary saved as 'Attrition_Summary.csv'")
```
